```python
import pandas as pd
data=pd.read_csv('weatherAUS.csv');
```

```python
data.info()
```

```python
data.dropna(subset=['RainTomorrow'],inplace=True)
```

```python
data['RainTomorrow'].isnull().sum()
```

```python
data.isnull().sum()
```

```python
numColumns=['MinTemp','MaxTemp','Rainfall','Evaporation','WindSpeed9am','WindSpeed3pm','Humidity9am','Humidity3pm','Pressure9am',
         'Pressure3pm','Cloud9am','Cloud3pm','Temp9am','Temp3pm','WindGustSpeed','Sunshine'     
]
numColumns
```

```python
for i in numColumns:
    data[i].fillna(data[i].mean(),inplace=True);
    
```

```python
data.isnull().sum()
```

```python
nomColumns=['WindGustDir','WindDir9am','WindDir3pm','RainToday']
nomColumns
```

```python
for i in nomColumns:
#     print(data[i].mode())
    data[i].fillna(data[i].mode(),inplace=True);
```

```python
data.isnull().sum()
```

```python
data.to_csv('weatherAUS.csv',index=False)
```
